# Sort an array stochastically

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_steps = 10000
n_elements = 109
history = np.zeros((n_steps, n_elements))

x = np.random.uniform(size=(n_elements))

for t in range(n_steps):
    history[t,:] = x
    i,j = (np.random.randint(n_elements, size=2).tolist())
    if i<j and x[i]>x[j]:
        x[i],x[j] = x[j],x[i]
    if i>j and x[i]<x[j]:
        x[i],x[j] = x[j],x[i]
    if np.random.uniform()<0.05:
        x[np.random.randint(n_elements)] = np.random.uniform()
    
plt.figure(figsize=(14,5))
plt.imshow(history.T, aspect='auto');
```
