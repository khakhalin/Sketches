# Read and analyze an image

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
```

```python
im = Image.open('../../images/purples.png')
im = im.convert('RGB')
# display(im) # Full image
y = np.array(im)

plt.figure(figsize=(12,6))
plt.plot(y[100,:,:]);
plt.legend(['r','g','b'], loc='right');
```
