# Fourier illustrations

Just some illustrations for a course. Of why Fourier transform is even possible.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Synthesis

a = np.array([8,6,4,2,1])
a = a/np.max(a)
x = np.linspace(0,4*np.pi,1000)
y = np.zeros(x.shape)

fig = plt.figure(figsize=(12,8));
fig.patch.set_facecolor('white')
ax = plt.subplot(2,1,1)
for i in range(len(a)):
    z = np.sin(x*i)*a[i]
    y += z
    plt.plot(x, z, alpha=0.5)
plt.plot(x, y, color='black', linewidth=2)
nodes = np.pi*np.arange(0,5)
plt.plot(nodes, np.zeros(nodes.shape), '.', markersize=15, color='black');
ax.axis('off');

# Deliberately bad synthesis
ax = plt.subplot(2,1,2)
for i in range(len(a)):
    z = np.sin(x*np.sqrt(i*2.12))*a[i]
    plt.plot(x, z, alpha=0.5)
plt.plot(x, y, color='black', linewidth=2)
ax.axis('off');
```
