# Jumping numbers

Numbers with consecutive digits differing by one. The task is to find all jumpint numbers below input N.

```python
def jumpers(n):
    bag = []
    for i in range(1,min(10,n)):
        bag.append(i)
    if n<=10: return bag
    left = 0 # Lefts point of last generation
    right = len(bag)
    while True:
        for i in range(left,right):
            if (bag[i] % 10) > 0:
                k = bag[i]*10+(bag[i] % 10) - 1
                if k>n: return bag
                bag.append(k)
            if (bag[i] % 10) < 9:
                k = bag[i]*10+(bag[i] % 10) + 1
                if k>n: return bag
                bag.append(k)
        left = right
        right = len(bag)

print(jumpers(100000))
```
