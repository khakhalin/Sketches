```python
def read_text():
    f = open("input3.txt","r",encoding='utf-8');
    mytext = f.readlines()
    return mytext

def input():
    return mytext.pop(0)
```

```python
"""
4
SSSEEE
N
N3(S)N2(E)N
2(3(NW)2(W2(EE)W))

Case #1: 4 4
Case #2: 1 1000000000
Case #3: 3 1
Case #4: 3 999999995
"""
```

```python
mytext = read_text() # The only line not to be copied

mem = {}
big = 1000000000
NUMS = {str(n) for n in range(2,10)}

def do(prog,xy):
    if prog in mem:
        d = mem[prog]
        return [xy[0]+d[0], xy[1]+d[1]]
    d = [0,0]
    i = 0
    while i<len(prog):
        s = prog[i]
        if s in NUMS:
            x = int(s)
            left = i+2
            i += 1
            level = 1
            while level>0:
                i += 1
                if prog[i]=='(': level += 1
                if prog[i]==')': level -= 1
            right = i
            #print('Subprogram:',prog[left:right])
            shift = do(prog[left:right],(0,0))
            d = [d[0]+x*shift[0], d[1]+x*shift[1]]
        else:
            if s=='N': d[1] -= 1
            if s=='S': d[1] += 1
            if s=='W': d[0] -= 1
            if s=='E': d[0] += 1
        i += 1
    mem[prog] = d
    return [xy[0]+d[0], xy[1]+d[1]]

n_cases = int(input())
for i_case in range(n_cases):
    prog = input().strip()
    #print(prog)
    xy = do(prog,(1,1))
    xy[0] = xy[0] % big
    xy[1] = xy[1] % big
    if xy[0]==0: xy[0] = big
    if xy[1]==0: xy[1] = big
    print("Case #" + str(i_case+1) + ": ",xy[0],xy[1])
```
