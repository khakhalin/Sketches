# Checking how sklearn ridge regresson works

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
```

```python
# Generate data
size = [100, 5]

x = np.zeros(size)
t = np.arange(size[0])/size[0]*2*np.pi
for i in range(size[1]):
    x[:,i] = np.sin(t * i**2)/(1+i**2) # Just some formula idk

plt.plot(x);
```

```python
c = np.random.normal(size=(size[1],1))
y = x @ c
y_noisy = y + np.random.normal(size=y.shape)
plt.plot(t, y_noisy, '.');
plt.plot(t, y);
```

```python
# Ridge regression

clf = lm.Ridge(alpha=1)
clf.fit(x, y_noisy)
c_hat = clf.coef_.T
print(c_hat)

y_hat = x @ c_hat
plt.plot(t, y_hat);
plt.plot(t, y, '.');
```
