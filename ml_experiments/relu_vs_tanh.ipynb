# A fight of ReLu and Tanh

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-3,5,0.1)

relu = lambda x: np.maximum(0,x)

plt.figure(figsize=(4,4))
plt.plot(x, relu(2*x + 1))           # One layer (w & b)
plt.plot(x, np.tanh(x/2-3/4)*4 + 4); # Have to engage next layer
```
