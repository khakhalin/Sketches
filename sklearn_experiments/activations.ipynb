# Playing with activation functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Playing with weird custom functions

x = np.linspace(-5, 5, 1000)

def y(x, a=10):
    return x*(1 + a/(1+abs(x)))

fig = plt.figure()
fig.patch.set_color('white')
plt.plot(x, y(x));
```

```python
# Playing with weird custom functions

x = np.linspace(-5, 5, 1000)

def y(x, a=10):
    return x*(1 + a/(1+np.square(x)))

fig = plt.figure()
fig.patch.set_color('white')
plt.plot(x, y(x));
```
